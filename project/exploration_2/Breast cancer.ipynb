# Expl. 2-3) Breast cancer wisconsin (diagnostic) dataset



```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

##### 1) 모델 학습, 수행

```python
cancer = load_breast_cancer()
data_cancer = cancer.data
label_cancer = cancer.target
```

# Breast cancer wisconsin (diagnostic) dataset

```python
print(cancer.keys())
```

```python
# 데이터 Describe 해 보기
print(cancer.DESCR)
```

##### 2) 데이터 분리 (train, test)

```python
x_train, x_test, y_train, y_test = train_test_split(data_cancer, label_cancer, test_size=0.2, random_state=1)

print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

##### 3) 모델 학습, 수행

### Decision Tree

```python
decision_tree = DecisionTreeClassifier(random_state=10)
decision_tree.fit(x_train, y_train)
y_pred = decision_tree.predict(x_test)
print(classification_report(y_test, y_pred))
```

### Random Forest


```python
random_forest = RandomForestClassifier(random_state=10)
random_forest.fit(x_train, y_train)
y_pred = random_forest.predict(x_test)
print(classification_report(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```

### SVM


```python
svm_model = svm.SVC()
svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)
print(classification_report(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```

### SGD Classifier


```python
sgd_model = SGDClassifier()
sgd_model.fit(x_train, y_train)
y_pred = sgd_model.predict(x_test)
print(classification_report(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```

### Logistic Regression

```python
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
y_pred = logistic_model.predict(x_test)
print(classification_report(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```
